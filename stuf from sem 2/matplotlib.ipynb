### Legends, Titles, and Labels

```python
years=[1950,1955,1960,1965,1970,1980,1985,1990,1995,2000,2005,2010,2015]
pops=[2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]
death=[1.2,1.1,1.2,2.1,2.0,2.3,1.8,1.9,2.6,1.6,2.4,2.4,4.0]
plt.plot(years,pops,color='b',label='pops')
plt.plot(years,death,'--',color='r',label='death')
plt.ylabel("Population in Billions")
plt.xlabel("Population growth by year")
plt.title("Population Growth")
plt.legend()
plt.show()
```

### A basic histogram on the sepal length of all iris flowers, using 20 equal-length bins

```python
# Extract the sepal length column from the dataset
sepal_length = iris_df['sepal_length']

# Create a histogram with 20 equal-length bins
plt.hist(sepal_length, bins=20)

# Add a title and axis labels
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

# Show the histogram
plt.show()
```

### Plot the histograms for all features by creating separate with use of subplots, so that all histograms are shown in one single plot 

```python
# Create a figure with subplots for each feature
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

# Plot a histogram for each feature in its corresponding subplot
axs[0, 0].hist(iris_df['sepal_length'], bins=20)
axs[0, 0].set_title('Sepal Length')

axs[0, 1].hist(iris_df['sepal_width'], bins=20)
axs[0, 1].set_title('Sepal Width')

axs[1, 0].hist(iris_df['petal_length'], bins=20)
axs[1, 0].set_title('Petal Length')

axs[1, 1].hist(iris_df['petal_width'], bins=20)
axs[1, 1].set_title('Petal Width')

# Add a main title for the whole plot
fig.suptitle('Histograms of Iris Features')

# Adjust the spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```
