```python
import matplotlib.pyplot as plt
```

```python
# Extract the columns
x = sal['YearsExperience']
y = sal['Salary']

# Create the scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')

# Show the plot
plt.show()
```

```python
colors = ['red' if exp > 5 else 'blue' for exp in x]

# Create the scatter plot
plt.scatter(x, y, c=colors)

# Add labels and title
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')

# Show the plot
plt.show()
```
