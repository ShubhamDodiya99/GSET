```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
# Bar plot of Gender
gender_counts = df["Gender"].value_counts()
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(gender_counts.index, gender_counts.values)
ax.set_xlabel("Gender")
ax.set_ylabel("Count")
ax.set_title("Distribution of Gender")
plt.show()
```

```python
# Histogram of Credit Score
fig, ax = plt.subplots(figsize=(6, 4))
ax.hist(df["CreditScore"], bins=20, color="blue")
ax.set_xlabel("Credit Score")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of Credit Scores")
plt.show()
```

```python
# Box plot of CreditScore by Geography
fig, ax = plt.subplots(figsize=(7, 5))
df.boxplot(column="CreditScore", by="Geography", ax=ax)
ax.set_xlabel("Geography")
ax.set_ylabel("Credit Score")
ax.set_title("Credit Score Distribution by Geography")
plt.show()
```

```python

# Pie chart of Exited customers
fig, ax = plt.subplots(figsize=(7, 5))
exit_counts = df["Exited"].value_counts()
ax.pie(exit_counts, labels=["Not Exited", "Exited"], autopct="%1.1f%%")
ax.set_title("Exited Customers Proportion")
plt.show()
```

```python

# Pie chart of Exited customers
fig, ax = plt.subplots(figsize=(7, 5))
exit_counts = df["Exited"].value_counts()
labels = ["Not Exited", "Exited"]
explode = (0.05, 0.05) # Explode the slices to highlight the exit count

ax.pie(exit_counts, labels=labels, explode=explode, autopct="%1.1f%%")
ax.set_title("Exited Customers Proportion")
plt.show()
```

```python
# Line chart of Estimated Salary by Age
fig, ax = plt.subplots(figsize=(7, 5))
df.groupby("Age")["EstimatedSalary"].mean().plot(ax=ax)
ax.set_xlabel("Age")
ax.set_ylabel("Estimated Salary")
ax.set_title("Estimated Salary by Age")
plt.show()
```

```python

# Violin plot of Age distribution
fig, ax = plt.subplots(figsize=(7, 5))
ax.violinplot(df["Age"], showmeans=True, showextrema=True)
ax.set_title("Age Distribution")
ax.set_ylabel("Age")
plt.show()
```

```python
# Area chart of Balance by Age
fig, ax = plt.subplots(figsize=(7, 5))
df.groupby("Age")["Balance"].mean().plot.area(ax=ax)
ax.set_xlabel("Age")
ax.set_ylabel("Balance")
ax.set_title("Balance by Age")
plt.show()
```

```python
# Scatter plot of Credit Score vs. Balance
fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(df["CreditScore"], df["Balance"])
ax.set_xlabel("Credit Score")
ax.set_ylabel("Balance")
ax.set_title("Credit Score vs. Balance Scatter Plot")
plt.show()
```

```python
# Heatmap of Correlation Matrix
fig, ax = plt.subplots(figsize=(7, 5))
corr = df.corr()
im = ax.imshow(corr, cmap="coolwarm")
ax.set_xticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=90, ha="right")
ax.set_yticks(range(len(corr.columns)))
ax.set_yticklabels(corr.columns)
fig.colorbar(im)
ax.set_title("Correlation Matrix Heatmap")
plt.show()
```

```python
# 3D scatter plot of Age, CreditScore, and Balance
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df["Age"], df["CreditScore"], df["Balance"])
ax.set_xlabel("Age")
ax.set_ylabel("Credit Score")
ax.set_zlabel("Balance")
ax.set_title("Age vs. Credit Score vs. Balance")
plt.show()
```
