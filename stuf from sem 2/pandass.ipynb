```python
# Set the 'Name' column as the index
df.set_index('Name', inplace=True)
df.head()
```

```python
# Reset the index to the default integer index
df.reset_index(inplace=True)
df.head()
```

```python
# Sort the DataFrame by age in ascending order
df.sort_values('Age', inplace=True)
df.head()
```

```python
# Sort the DataFrame by age in descending order
df.sort_values('Age', ascending=False, inplace=True)
df.head()
```

```python
# Get the mean salary
mean_salary = df['Salary'].mean()
mean_salary
```

```python
# Get the standard deviation of salary
std_salary = df['Salary'].std()
std_salary
```

```python
# Get the minimum and maximum ages
min_age = df['Age'].min()
max_age = df['Age'].max()
print('minimum age:',min_age)
print('maximum age:',max_age)
```

```python
# Get the value counts of each department
dept_counts = df['Department'].value_counts()
dept_counts
```

```python
# Group the data by department and get the total salary for each department
salary_by_dept = df.groupby('Department')['Salary'].sum()
salary_by_dept
```

```python
# Group the data by gender and get the average age for each gender
age_by_gender = df.groupby('Gender')['Age'].mean()
age_by_gender
```

```python
# Group the data by department and get the number of employees in each department
count_by_dept = df.groupby('Department')['Name'].count()
count_by_dept
```

```python
# Group the data by department
grouped = df.groupby('Department')
```

```python
# Get the first row for each group
first_rows = grouped.first()
first_rows
```

```python
# Get the last row for each group
last_rows = grouped.last()
last_rows
```

```python
# Get the sum of salaries for each group
salary_sums = grouped['Salary'].sum()
salary_sums
```
