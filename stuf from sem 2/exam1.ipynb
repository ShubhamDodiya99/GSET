### Create dataframe from Dictionary

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
        'age': [25, 30, 35, 40],
        'gender': ['F', 'M', 'M', 'M']}

df = pd.DataFrame(data)

print(df)
```

### Indexing a dataframe

```python
# Select the row with index 1 and the 'name' column
print(df.loc[1, 'name'])  # Output: 'Bob'

# Select all rows up to and including index 2, and the 'age' and 'gender' columns
print(df.loc[:2, ['age', 'gender']])
```

### Fill missing values

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
        'age': [25, np.nan, 35, 40],
        'gender': ['F', 'M', np.nan, 'M']}

df = pd.DataFrame(data)

# Fill missing values with a constant value
df.fillna(0, inplace=True)

print(df)
```

### Replace missing value with 0

```python
# Replace all occurrences of np.nan with 0
df.replace(np.nan, 0, inplace=True)

print(df)
```

### Scatter plot using Matplot

```python
# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Create a scatter plot
plt.scatter(diamonds['carat'], diamonds['price'])
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Carat vs. Price')

# Display the scatter plot
plt.show()
```

### Histogram and Subplot with Matplot

```python
# Create a histogram subplot
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].hist(diamonds['carat'], bins=20, alpha=0.5, label='Carat')
axs[0].hist(diamonds['price'], bins=20, alpha=0.5, label='Price')
axs[0].legend(loc='upper right')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Carat and Price')

axs[1].hist(diamonds['depth'], bins=20, alpha=0.5, label='Depth')
axs[1].hist(diamonds['table'], bins=20, alpha=0.5, label='Table')
axs[1].legend(loc='upper right')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Depth and Table')

# Add a title to the entire figure
fig.suptitle('Diamonds Dataset')

# Display the histogram subplot
plt.show()
```

### Heatmap with Seaborn

```python
# Create a heatmap of the correlations between numeric variables
sns.heatmap(diamonds.corr(), annot=True, cmap='coolwarm')
plt.title('Correlations between numeric variables in Diamonds dataset')

# Display the heatmap
plt.show()
```

### Scatterplot with Seaborn

```python
# Create a scatter plot of carat versus price
sns.scatterplot(x='carat', y='price', data=diamonds)
plt.title('Scatter plot of carat versus price')

# Display the scatter plot
plt.show()
```

### Box plot with Seaborn

```python
# Create a boxplot of price by cut
sns.boxplot(x='cut', y='price', data=diamonds)
plt.title('Boxplot of price by cut')

# Display the boxplot
plt.show()
```

### Pie chart with Seaborn

```python
# Create a pie chart of the distribution of cuts
diamonds['cut'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of cuts in Diamonds dataset')

# Display the pie chart
plt.show()
```

### Joint plot with Seaborn

```python
# Create a joint plot of carat versus price
sns.jointplot(x='carat', y='price', data=diamonds)
plt.suptitle('Joint plot of carat versus price')

# Display the joint plot
plt.show()
```
